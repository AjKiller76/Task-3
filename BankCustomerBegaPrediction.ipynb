```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df.drop(axis=1,columns=['RowNumber','Surname'],inplace=True)
```

```python
df.drop(axis=1,columns=['CustomerId'],inplace=True)
```

```python
X=df.iloc[:,:-1].values
```

```python
y=df.iloc[:,-1].values
```

```python
X
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1,2])],remainder='passthrough')
X=np.array(ct.fit_transform(X))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=7)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
for i in range(1,100):
  classifier=RandomForestClassifier(n_estimators=i,criterion='entropy',random_state=0)
  classifier.fit(X_train,y_train)
  y_pred=classifier.predict(X_test)
  print(i,accuracy_score(y_test,y_pred))
```

```python
classifier=RandomForestClassifier(n_estimators=40,criterion='entropy',random_state=0)
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
